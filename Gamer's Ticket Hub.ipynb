```python
from datetime import datetime
import random
def greeting():
    print('Welcome to the Gamer Season - Tournaments Section.\n')
    print("The Gamer Season is the largest gathering of games and esports in the world, where elite esports players and gaming \n enthusiasts from around the world come together to compete for the strongest esports tournaments and discover the mysteries behind the creation of many electronic games. The world's strongest clubs gather in the heart of the Kingdom of Saudi Arabia in the capital Riyadh for eight weeks from July to September, where champions compete to reach the greatest summit of esports in the world.")  
    
   

cart = {
    'Name':[],
    'Date':[],
    'Price':[],
    'Type':[]
}

global total_price

def cart_management():
    # It shows system available features to manage user cart
    print('Welcome to cart management section, please check the following options \n')
    print('1- Print cart elements\n')
    print('2- Add cart elements\n')
    print('3- delete cart elements\n')
    choice = int(input('Enter your choice'))
    if choice == 1:
        print_cart_elements(cart)     
    elif choice == 2:
        tournaments_options()
    elif choice == 3:
        delete_from_cart(cart)

def print_cart_elements(cart_elements):
    total_price = 0
    if(len(cart['Name']) == 0): #checks if cart is empty 
        choice = int(input('You have no items, write 1 to proceed to tournaments section or 0 to exit code'))
        if(choice == 1):
            tournaments_options()
        else:
            print('Thank you for your visit')
            return
    else:
        total_price = 0
        print('\nHere is your cart information\n')
        for name,date,price,Type in zip(cart_elements['Name'],cart_elements['Date'],cart_elements['Price'],cart_elements['Type']):
            print(f'Name : {name}')
            print(f'Date : {date}')
            print(f'price : {price}')
            print(f'Type: {Type}\n')
            total_price+=price
        choice = int(input("'What would you like to do next? Type '1' for Payment or '2' for Main Menu:'"))
        if(choice == 1 ):
            total = total_price
            return total
        else:
            cart_management()


def add_to_cart(cart_elements):
    choice = int(input('Would you like to book a ticket? please write 1 to yes and 0 to no'))
    while choice != 0:
        if(choice == 1):
            tournament_name = str(input('Please enter the name of tournament'))
            for name,date,price,Type in zip(cart_elements['Name'],cart_elements['Date'],cart_elements['Price'],cart_elements['Type']):
                if(tournament_name == name):
                    cart_choice = int(input('Add this ticket to your cart? Type 1 for Yes, 0 for No'))
                    if(cart_choice == 1):
                        cart['Name'].append(name)
                        cart['Date'].append(date)
                        cart['Price'].append(price)
                        cart['Type'].append(Type)
                        choice = int(input('Book another ticket? Type 1 for Yes, 0 for No:'))
                        if(choice == 1):
                            continue
                        else:
                            cart_management()
def delete_from_cart(cart_elements):
    
    #shows to the user his available cart elements and allows them to delete ticket by entering ticket's name
    if(len(cart['Name']) == 0): #checks if cart is empty 
        choice = int(input('You have no items, write 1 to proceed to tournaments section or 0 to exit code'))
        if(choice == 1):
            tournaments_options()
        else:
            print('Thank you for your visit')
            return
    else:
        print('\nHere is your cart information\n')
        for name,date,price,Type in zip(cart_elements['Name'],cart_elements['Date'],cart_elements['Price'],cart_elements['Type']):
            print(f'Name : {name}')
            print(f'Date : {date}')
            print(f'price : {price}')
            print(f'Type: {Type}\n')
        print('To delete a ticket, type its name and press Enter:')
        ticket_name = str(input('Ticket name : '))
        for name,date,price,Type in zip(cart_elements['Name'],cart_elements['Date'],cart_elements['Price'],cart_elements['Type']):
            if(name == ticket_name):
                cart['Name'].remove(name)
                cart['Price'].remove(price)
                cart['Date'].remove(date)
                cart['Type'].remove(Type)
        cart_management()
        

def Elite_Tournaments():
    
    print('\n" Elite Tournaments "\n')
    
    EliteTournaments = {
    'Name':['Riyadh Masters','ESL','Tekken','CS GO'],
    'summary':["The Riyadh Masters in Dota2 is a key part of the Gamers8 Festival. Over a period of two weeks, from July 17th to July 30th, twenty of the world's strongest teams compete for a sizeable prize pool of $15,000,000 USD. The tournament features a play-in, group stages, and double-elimination playoffs. The event is considered a breeding ground for champions.","The ESL R1 and Sim Racing Title Rennsport tournaments have joined Gamers8, with 12 teams and 48 players competing for a $1,000,000 prize pool. The tournaments offer a new digital racing experience that blends traditional racing culture with modern technology. The four-day event promises to be full of exciting racing.","Tekken 7 is making its debut at Gamers8, with 16 national teams from around the world competing for a total prize pool of $1,000,000 USD. The tournament spans over four days, featuring brutal rivalry with over 40 characters to choose from and intense gameplay mechanics such as opponent juggling, rage arts, and crazy whiff punishes. The Nations Cup presents a new tournament system to ensure that only the most deserving team emerges victorious.","Gamers8 Festival is hosting a Counter-Strike: Global Offensive tournament from August 16th to 30th. This tournament is one of the last few before CS:GO is replaced with Counter-Strike 2. Sixteen top-tier teams will compete for a share of the impressive $1,000,000 prize pool. Fans should mark their calendars and get ready for some intense gameplay."],
    'Date':[datetime(2023, 8, 19, 10, 30, 0).strftime("%Y-%m-%d %I:%M:%S %p"),datetime(2023, 8, 28, 10, 30, 0).strftime("%Y-%m-%d %I:%M:%S %p"),datetime(2023, 9, 28, 10, 30, 0).strftime("%Y-%m-%d %I:%M:%S %p"),datetime(2023, 9, 28, 10, 30, 0).strftime("%Y-%m-%d %I:%M:%S %p")],
    'Price':[30,35,40,45],
    'Type':['Elite','Elite','Elite','Elite']
}

    for name,summary,date,price in zip(EliteTournaments['Name'],EliteTournaments['summary'],EliteTournaments['Date'],EliteTournaments['Price']):
        print(f'\nTournament Name: {name} \n')
        print(f'Tournament summary : \n {summary.upper()}')
        print(f'Date :  {date}'),
        print(f'Ticket Price  : {price} SR')
        print('\n*************************\n')
    
    add_to_cart(EliteTournaments)        
        

        

    
def Community_Tournaments():
    
    print('\n" Community Tournaments "\n')
    
    CommunityTournaments = {
    'Name':['Rocket League','Street Fighter 6','FIFA 23'],
    'Date':[datetime(2023, 8, 3, 22, 45, 0).strftime("%Y-%m-%d %I:%M:%S %p"),datetime(2023, 8, 3, 20, 10, 0).strftime("%Y-%m-%d %I:%M:%S %p"),datetime(2023, 8, 3, 19, 34, 0).strftime("%Y-%m-%d %I:%M:%S %p")],
    'Price':[10,10,20],
    'Type':['Community','Community','Community','Community']
}
    for name,date,price in zip(CommunityTournaments['Name'],CommunityTournaments['Date'],CommunityTournaments['Price']):
        print(f'\nTournament Name: {name} \n')
        print(f'Date :  {date}'),
        print(f'Ticket Price  : {price} SR')
        print(f'Location : Challenge Zone')
        print('\n*************************\n')
        
    add_to_cart(CommunityTournaments)



def intl_reg_tournaments(): #International/Regional tournaments
    intl_reg_Tournaments = {
    'Name':['Dota 2','FIFA 23','Valorant'],
    'Summary':['KSA IS HOSTING A FRIENDLY MATCH BETWEEN THAILAND NATIONAL TEAM AND KSA NATIONAL TEAM','KSA IS HOSTING THE AFRICAN AND ARAB FEDERATIONS TO PARTICIPATE AFRO - ARAB ESPORTS LEAGUE IN \n GAMERS 8','KSA IS HOSTING THE AFRICAN AND ARAB FEDERATIONS TO PARTICIPATE AFRO - ARAB ESPORTS LEAGUE IN \n GAMERS 8'],
    'Date':[datetime(2023, 9, 19, 20, 45, 0).strftime("%Y-%m-%d %I:%M:%S %p"),datetime(2023, 9, 21, 20, 10, 0).strftime("%Y-%m-%d %I:%M:%S %p"),datetime(2023, 9, 19, 19, 34, 0).strftime("%Y-%m-%d %I:%M:%S %p")],
    'Competitors':['KSA VS THAILAND','Arab and African federations','Arab and African federations'],
    'Price':[100,150,200],
    'Type':['International','Regional','International']
    }
        
    for name,summary,date,competitors,price in zip(intl_reg_Tournaments['Name'],intl_reg_Tournaments['Summary'],intl_reg_Tournaments['Date'],intl_reg_Tournaments['Competitors'],intl_reg_Tournaments['Price']):
        print(f'\nTournament Name: {name} \n')
        print(f'Tournament summary :  {summary.upper()}')
        print(f'Date :  {date}'),
        if price == 0:
            print('Sold out')
            continue
        print(f'Competitors : {competitors}')
        print(f'Ticket Price  : {price} SR')
        print(f'Location : Challenge Zone')
        print('\n*************************\n')
        
    add_to_cart(intl_reg_Tournaments)

def tournaments_options():
    print('\nIn Tournaments section we cover three different types of tournaments which one would like to see? ')

    print('\n1-Elite Tournaments')
    print('\n2-Community Tournaments')
    print('\n3-International/Regional Tournaments\n')
    
    while True:
        choice = int(input())
        if(choice == 1):
            Elite_Tournaments()
            break
        elif(choice == 2):
            Community_Tournaments()
            break
        elif(choice == 3):
            intl_reg_tournaments()
            break
        else:
            print('This option is not available please try again')
```
